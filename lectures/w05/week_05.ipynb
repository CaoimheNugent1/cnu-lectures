# Matplotlib (object-oriented plotting)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,1000)
c = lambda n: np.cos(np.pi*n*x)
s = lambda n: np.sin(np.pi*n*x)
```

```python
n = 5
plt.plot(x,c(n),lw=3,color='purple')
plt.xlabel('x');
plt.title(f'cos({n}Ï€x)');
plt.savefig('test.png')
```

```python
fig, ax = plt.subplots(2, 4, sharex=True, sharey=True)

for j in range(4):
    ax[0,j].plot(x,c(j),color='blue',lw=2.5)
    if j == 0:
        ax[0,j].set_title(f'1')
    elif j == 1:
            ax[0,j].set_title(f'cos(Ï€x)')
    else:
        ax[0,j].set_title(f'cos({j}Ï€x)')
    
    ax[1,j].plot(x,s(j+1),color='red',lw=2.5)
    if j == 0:
        ax[1,j].set_title(f'sin(Ï€x)')
    else:
        ax[1,j].set_title(f'sin({(j+1)}Ï€x)')
    ax[1,j].set_xlabel('x')
```

The point here is that these are clear enough to convey information without being difficult to interpret. We could definitely make them better, but this is pretty ok for an interactive demo. You might notice that sine and cosine resemble each other more than you might first think when plotted this way. They are roughly positive and negative in the same locations for the corresponding plots. 

That's the point of a good plot -- make something clear that wasn't before. 

At this point, it might be useful to see some of more polished examples of things matplotlib can do. These are not easy to make, but that's ok if you're going to use it for something important.

### A few examples to experiment with

#### Making an x

##### `np.linspace(a,b,n)`

This gives you an `array` 

* Starting ***exactly*** at `a` 

* Ending ***exactly*** at `b`

* Containing ***exactly*** `n` values.


##### `np.arange(a,b,dx)`

This gives you an `array` 

* Starting ***exactly*** at `a` 

* Ending ***before*** at `b`

* With spacing between the values of `dx`


```python
a,b = 0.2, 1.5

n = 5
dx = 1/5

x1 = np.linspace(a,b,n)
x2 = np.arange(a,b,dx)

print(len(x1))
print(len(x2))

print(x1)
print(x2)
```

#### Single `plots`

##### Here is the simplest kind of plot you can make 

It contains no labels, and there is no adjustment to the lines and colors etc.

```python
x = np.linspace(0,1,1000)
y = np.sin(np.pi*x)
plt.plot(x,y);
```

#### We can put multiple plots together easily

```python
plt.plot(x,y)
plt.plot(x,y**10)
plt.plot(x,y**0.1)
```

### Graphics are objects.  

This means that all figures have `methods` that we can use to make more thing happen. 

Often, the method have `keywords` that can help make them look nicer.

##### You can even use `Markdown` in your labels

```python
# This changes the figure size
plt.figure(figsize=(8,6))

plt.plot(x,y,     linewidth=2.3,color='blue')
plt.plot(x,y**10, linewidth=2.3,color='orange')
plt.plot(x,y**0.1,linewidth=2.3,color='green')


# The x-axis
plt.xlabel('This is the horizontal axis (x)',fontsize=15)

# The y-axis
plt.ylabel('This is the vertical axis (y)',fontsize=15)

# The y-axis
plt.title('This is the BOLD Markdown title:    $y=sin(\\pi\, x)$',fontsize=15,fontweight='bold');

# The legend
plt.legend(['y','$y^{10}$','$y^{1/10}$'],fontsize=15)

# This alters the numbers on the axes.
plt.tick_params(axis='both',labelsize=15)

# This could include a big think x-axis marker
#plt.axhline(linewidth=4, color="black")

# This could include a big think y-axis marker
#plt.axvline(linewidth=4, color="black")

# This saves the figure to the local directory
plt.savefig('sample_plot.png')
```

For a 2 by 4 grid:

This has some extra options. For example, we can take away the redundant labels.  

```python
fig, ax = plt.subplots(2, 4, sharex=True, sharey=True)

for j in range(4):
    ax[0,j].plot(x,np.sin((j+1)*np.pi*x),linewidth=2,color='red')
    ax[1,j].plot(x,np.cos(j*np.pi*x),linewidth=2,color='blue')
    ax[0,j].set_ylim(-2, 2)
    ax[1,j].set_ylim(-2, 2)
    
ax[0,0].set_title("$\sin(\pi x)$")
ax[0,1].set_title("$\sin(2\pi x)$")
ax[1,0].set_title("$1$")
ax[1,1].set_title("$\cos(\pi x)$")    
for j in range(2,4): 
    ax[0,j].set_title("$\sin(%i \pi x)$" %(j+1))
    ax[1,j].set_title("$\cos(%i \pi x)$" %(j))


for i in range(2):
    ax[i,0].set_ylabel('$y(x)$',fontsize=15)
    
for j in range(4):
    ax[1,j].set_xlabel('$x$',fontsize=15)
```

#### Lots of different kinds of plots:

#### **HISTOGRAMS:** 
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

Play around with the options. `alpha` changes the color opacity. You can see that `'red'` + `'blue'` = `'purple'`. This wouldn't be the case with other values of `alpha`.

```python
fig, ax = plt.subplots()

x = np.random.normal(size=10000)
H = ax.hist(x, bins=50, alpha=1.0, histtype='stepfilled',color='black')

x = np.random.normal(size=5000)
H = ax.hist(x, bins=50, alpha=0.6, histtype='stepfilled',color='red')

x = np.random.normal(size=2000)
H = ax.hist(x, bins=50, alpha=1.0, histtype='stepfilled',color='blue')
```

##### **ERROR BARS** https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html

```python
x = np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots()
ax.errorbar(x, y, dy, fmt='.k')
ax.plot(x, y)
```

##### **SCATTERPLOT** https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html

```python
x = 2*np.random.random(50)
y = x**3+2*np.random.random(50)
c = 0.2*x+0.5*x*np.random.random(50)          # color of points
s = x+ 200*np.random.random(50)    # size of points

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap=plt.cm.jet)

# Add a colorbar
fig.colorbar(im, ax=ax)

# set the color limits - not necessary here, but good to know how.
im.set_clim(0.0, 1.0)

ax.set_xlabel('Known Variable',fontsize=15)
ax.set_ylabel('Measured Variable',fontsize=15);

ax.set_title('There seems to be a correlation',fontsize=15);

ax.tick_params(axis='both',labelsize=15)
```

#### CONTOUR PLOTS

#### Broadcasting

We can first see a useful trick for making 2D arrays from 1D arrays.

Here is a 1D array

```python
x = np.linspace(0, 10*np.pi, 100)

print(x.shape)
```

What if we put in a fake axis?

```python
print('Make a column vector:')
print('')

print(x[0:5,None])

print('')
print('Make a row vector:')

print('')
print(x[None,0:5])

print('')


print('Add the two together, get a 2D array:')
print('')
print(x[0:5,None]+x[None,0:5])
```

Now we can combine them like they are 2D arrays.

Lets make the function 

$$
\large z(x,y) \ = \ \left[\,\cos(x) + 2 \cos \!\left(\!\frac{x}{2}\!\right)\,\right]\, \cos\!\left(\!\!\!\frac{\sqrt{3} y}{2}\!\!\right)
$$

```python
x = np.linspace(0, 10*np.pi, 100)

x, y = x[:,None], x[None,:]

z = (np.cos(x) + 2*np.cos(x/2)*np.cos(np.sqrt(3)*y/2))


x, y = x+0*y, y+0*x


print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# contour lines
im0 = ax.contour(x, y,-z,(0.7,0.8,0.9,1.0,1.1,1.2,1.3,1.4),colors='k')

# filled contours
im1 = ax.contourf(x, y, -z, 10 )

ax.set_xlabel('x',fontsize=20)
ax.set_ylabel('y',fontsize=20)

ax.tick_params(axis='both',labelsize=20)

# Use these if you want a colorbar (I don't right now)
#fig.colorbar(im1, ax=ax)
#ax.set_aspect('equal')

# save the figure
fig.savefig('hexagons.png')
```

### **2D COLOR PLOTS** 

### `imshow`

Here is an example of `imshow` https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html

```python
I = np.random.random((200, 200))

I += np.tanh(np.linspace(-1, 5, 200))

fig, ax = plt.subplots(figsize=(7, 7))

im = ax.imshow(I, cmap=plt.cm.jet)

fig.colorbar(im, ax=ax)
ax.set_aspect('equal')
```

### similar result with `pcolormesh`

Here is another example using ``pcolormesh``, which needs a grid of $x$ and $y$ values.
https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.pcolormesh.html

The good thing about `pcolormesh` is that it can work for data on a non-uniform grid.

```python
I = np.random.random((200, 200))
x = np.logspace(-2,-1,200)
y = np.arange(200)

I += np.tanh(np.linspace(-1, 5, 200))

fig, ax = plt.subplots(figsize=(7, 7))

im = ax.pcolormesh(x,y,I, cmap=plt.cm.jet)

fig.colorbar(im, ax=ax)
# ax.set_aspect('equal')
```
